```python
import os

os.chdir('../data/')
DATA_PATH = os.getcwd()

print(DATA_PATH)
```

```python
RAW_DATA_PATH = DATA_PATH + "\scenario_raw"
READ_DATA_PATH = DATA_PATH + "\scenario_text"
data_list = os.listdir(RAW_DATA_PATH)
print(data_list)
```

```python
import sys
import subprocess

# 필요한 라이브러리의 이름을 리스트로 나열합니다.
#  pdfplumber, pdfminer.six, PyPDF2,PyMuPDF, tabula
required_libraries = ['PyMuPDF', 'konlpy', 'soynlp']

# 현재 설치된 라이브러리 목록 얻기
installed_libraries = [package.split(' ')[0] for package in subprocess.check_output([sys.executable, '-m', 'pip', 'list']).decode('utf-8').split('\r\n')[2:]]

# 필요한 라이브러리를 확인하고 필요한 경우 설치합니다.
for library in required_libraries:
    print(library, '확인 중')
    if library not in installed_libraries:
        try:
            print('설치 시작')
            subprocess.check_call([sys.executable, '-m', 'pip', 'install', library])
        except subprocess.CalledProcessError:
            print(f"라이브러리 {library} 설치 실패.")
    else:
        print(f"{library} 라이브러리가 이미 설치되어 있습니다.")
```

```python
import fitz, pathlib
for script in data_list:
    print(script, '시작')
    file_path = os.path.join(RAW_DATA_PATH, script)
    print(file_path)
    if script.endswith('pdf'):
        with fitz.open(file_path) as doc:  # open document
            text = chr(12).join([page.get_text() for page in doc])
            pathlib.Path(READ_DATA_PATH + '\\' + script + ".txt").write_bytes(text.encode())
    else:
        print('다른 데이터 타입 개발 필요')
    print(script, '종료')
```

```python
nlp_raw_list = os.listdir(READ_DATA_PATH)
print(nlp_raw_list)

raw_file_path = os.path.join(READ_DATA_PATH, nlp_raw_list[0])
with open(raw_file_path, 'r') as f:
    raw_file = f.read()
```

```python
import konlpy
from konlpy.tag import Kkma, Okt, Hannanum, Komoran

#형태소 분석기
kkma = Kkma()
okt = Okt()

kon_morphs = okt.morphs(raw_file, stem= True)
kon_nouns = okt.nouns(raw_file)
kon_phrases = okt.phrases(raw_file)
kon_pos = okt.pos(raw_file)
```

```python
print(kon_nouns)
```

```python
from collections import Counter
import csv

for i, v in enumerate(kon_nouns):
    print(i, v)
    if len(v) < 2:
        print(v)
        # kon_nouns.pop(1)

count = Counter(kon_nouns)

kon_noun_list = count.most_common(100)
for v in kon_noun_list:
    print(v)
    
    
# txt 파일에 저장
with open("kon_noun_list.txt",'w',encoding='utf-8') as f:
	for v in kon_noun_list:
		f.write(" ".join(map(str,v))) #튜플 int값을 str 타입으로 전환 후 조인
		f.write("\n")

# csv 파일에 저장
with open("noun_list.csv","w", newline='',encoding='euc-kr') as f:
	csvw = csv.writer(f)
	for v in kon_noun_list:
		csvw.writerow(v)
```

```python
from soynlp.utils import DoublespaceLineCorpus
from soynlp.word import WordExtractor
from soynlp.noun import LRNounExtractor_v2

raw_file_path = os.path.join(READ_DATA_PATH, nlp_raw_list[0])

train_data = DoublespaceLineCorpus(raw_file_path) # 데이터기반패턴학습

noun_extractor = LRNounExtractor_v2()
nouns = noun_extractor.train_extract(train_data) # [할리우드, 모바일게임 ...
```

```python
list(noun_extractor._compounds_components.items())
```

```python
noun_extractor.lrgraph.get_r('친구')
```

```python
noun_extractor.lrgraph.get_l('를')
```

```python
word_extractor = WordExtractor(min_frequency=100,
    min_cohesion_forward=0.05, 
    min_right_branching_entropy=0.0
)
word_extractor.train(train_data) # [클린턴, 트럼프, 프로그램...
words = word_extractor.extract()

cohesion_score = {word:score.cohesion_forward for word, score in words.items()}
noun_scores = {noun:score.score for noun, score in nouns.items()}
combined_scores = {noun:score + cohesion_score.get(noun, 0)
    for noun, score in noun_scores.items()}
combined_scores.update(
    {subword:cohesion for subword, cohesion in cohesion_score.items()
    if not (subword in combined_scores)}
)
```

```python
from soynlp.tokenizer import LTokenizer

tokenizer = LTokenizer(scores=cohesion_score)
print(tokenizer.tokenize(raw_file, flatten=True))
```

```python
from soynlp.tokenizer import MaxScoreTokenizer

tokenizer = MaxScoreTokenizer(scores=cohesion_score)

print(tokenizer.tokenize('난파스타가 좋아요', flatten=False))
```

```python
from soynlp.tokenizer import RegexTokenizer

tokenizer = RegexTokenizer()

print(tokenizer.tokenize(raw_file))
```
